###REGRESI LINEAR

regresi linear = metode statistik yang melakukan prediksi dengan variabel target berupa numerik, melihat korelasi linear antara dua atau lebih variabel, melihat bagaimana variansi dari suatu variabel dapat berpengaruh ke variansi variabel yang lain dengan kata lain melakukan prediksi terhadap suatu nilai dari variabel y berdasarkan data dari variabel x.

###Membangun model regresi linear

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
x= data[['total_bill']]
y= data[['tip']]
```

```python
plt.scatter(x,y)
```

Data preparation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

Membangun modelling

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

#training data
model.fit(x_train, y_train)
LinearRegression()
```

```python
pred = model.predict(x_test)
```

```python
print('y=' + str(model.coef_)+ 'x' + '+' + str(model.intercept_))
```

Membuat visualisasi

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), c='red', label = 'Regresi Linear')
plt.legend()
```

**Poisson regression**

Poisson regresion = salah satu regresi yang digunakan untuk mendapatkan model dari variabel dependen(y) atau prediktor, dan variabel independen(x) atau respon, dengan mengasumsikan bahwa variabel respon memiliki distribusi poisson

Distribusi poisson = salah satu bentuk distribusi dikrit yang menggambarkan jumlah dari suatu kejadian dalam satu satuan waktu atau wilayah atau sejenisnya. contoh jumlah mahasiswa yang lulus dalam satu tahun, jumlah pengunjung dalam sehari

distribusi poisson memberikan asumsi bahwa antar kejadian tidak saling berhubungan satu sama lain atau independen dan rate dari tiap kejadian bernilai konstan atau tetap

bentuk distribusi poison bisa berubah-ubah tergantung nilai mean nya, hal ini dikarenakan mean memiliki nilai yang sama dengan varianceScreenshot 2023-08-25 010508.png



cara kerja
model regresi poison biasa digunakan untuk memodelkan data yang berbentuk jumlah. Data yang dimaksud ialah data diskrit yang hanya bernilai positif. Data jumlah biasa digunakan untuk menghitung suatu kejadian dalam suatu interval waktu seperti per tahun, per satuan luas wilayah. contoh jumlah pasien kanker per tahun, jumlah kecelakaan lalu lintas dijalan tol per tahun, jumlah sumur resapan di suatu wilayah di Jakarta

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

url = 'http://www.stat.columbia.edu/~gelman/arm/examples/police/frisk_with_noise.dat'
df = pd.read_csv(url, skiprows=6, delimiter=' ')
df.head()
```

Screenshot 2023-08-25 191758.png

```python
x = (df.groupby(['eth', 'precinct'])[['stops', 'past.arrests']].sum().reset_index().pipe(pd.get_dummies, columns=['eth', 'precinct']).sort_values(by='stops').reset_index(drop=True)

     )#mengurutkan kolom eth dan precinct dengan groupby dan melihat data stops dan past.arrests secara jumlah. kemudian dilakukan get dummies untuk one hot encoding dari kolom eth dan precinct(karena bukan ordinal melainkan kategori) dan hilangkan variabel stop untuk variabel x dan ditaruh variabel y karena variabel dependen
y = x.pop('stops')
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**Modelling**

```python
from sklearn.linear_model import PoissonRegressor
```

```python
model = PoissonRegressor()
#training data
model.fit(x_train, y_train)
```

**Evaluation**

```python
pred_pr = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print('MAE :', mean_absolute_error(y_test, pred_pr))
print ('MSE :', np.sqrt(mean_squared_error(y_test, pred_pr)))
```
